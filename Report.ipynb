## <u>Part 2 : Multi-Class Article Classification:</u>

<p>In this project, we build a Data Classifcation pipeline for News Articles Collected from New York Times. </p>
<p>For the purpose of analysis we collect articles belonging to the following classification.<br />
1. Sports <br />
2. Travel<br />
3. Technology<br />
4. Business <br />
    </p>

### Block Diagram of the Data Classification pipeline:



This Article Classification pipeline can be classified into multiple steps,

### Step 1: Data Collection (32,000 Articles Collected)

Collected articles using Article Search and Archieve API of NYTimes. We were able to collect upward 32k Articles belonging to 6 classifications. We picked the 4 classifcations and proceed with feature extraction

### Step 2: Top Features

Ran wordcount map reduce algorithm on Spark and found the top words for each class. Performed stemming and stop words suppression for the words in the article. This will help us obtain the top-words per class.


### Step 3: Feature DataSet

Using all the top features obtained for the classes, create a dataset for all the articles collected.  

<b>Model:</b> <br />
class_value feat_1:feat_1_value feat_2:feat_2_value  ..... feat_n:feat_n_value <br/>

<b>Example:</b><br />
0 1:23 2:3 3:0 4:0 5:3<br />
1 1:2  2:0 3:33 4:0 5:23<br />
2 1:0  2:0 3:23 4:22 5:2<br />

The above is an example for 3 samples where,each article belongs to a classification.<br />
The interpretation is first sample belongs to Class 0 with Feature values 23,3,0,0,3 for features 1 through 5.<br />

### Step 4: ML Train and Classify

Used the dataset and trained Naive Bayes and Neural-Network based classifier. A new article from one of the 4 classifications can be classified using the model.
